```python
for i in range(5):
    for j in range(5):
        print("*", end = " ")
    print()
```

```python
for i in range(1, 6, 1):
    for j in range(i):
        print("*", end = " ")
    print()
```

```python
for i in range(1, 2*5, 1):
    stars = i
    if stars > 5:
        stars = 2*5 - i
    for j in range(stars):
        print("*", end = " ")
    print()
```

```python
for i in range(5):
    for j in range(5 - i):
        print(" ", end = " ")
    for j in range(2*i + 1):
        print("*", end = " ")
    for j in range(5 - i):
        print(" ", end = " ")
    print()
        
```

```python
for i in range(5, 0, -1):
    for j in range(5 - i):
        print(" ", end = " ")
    for j in range(2*i - 1):
        print("*", end = " ")
    for j in range(5 - i):
        print(" ", end = " ")
    print()
```

```python
size = 5
for i in range(size):
    spaces = size - i - 1
    stars = 2 * i + 1
    print(" " *spaces + "*"*stars)
```

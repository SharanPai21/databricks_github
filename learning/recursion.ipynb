```python
def rec(n):
    if n == 0:
        return 1
    return n*rec(n-1)

print(rec(4))
```

```python
def rec2(nums):
    if not nums:
        return 0
    return nums[0] + rec2(nums[1:])

print(rec2([1,2,3,4,5]))
```

```python
def rev(arr, left = 0, right = None):
    if right is None:
        right = len(arr) - 1

    if right < left:
        return
    
    arr[left], arr[right] = arr[right], arr[left]
    rev(arr, left + 1, right - 1)
    return arr

print(rev([1,2,3,4,5]))

```

```python
def pal(st, left = 0, right = None):
    if right is None:
        right = len(st) - 1


    if left < right:
        if st[left] == st[right]:
            return pal(st, left + 1, right - 1)
        else:
            return "not pal"
    else:
        return "Pal"
    

pal('aabbaa')
```
